```python
import feedparser
from newspaper import Article  # Correct import for newspaper3k


def scrape_news_from_feed(feed_url):
    """
    Fetch articles from an RSS feed URL and extract detailed information.
    """
    articles = []
    # Parse the RSS feed
    feed = feedparser.parse(feed_url)
    
    for entry in feed.entries:
        try:
            # Create a newspaper3k article object for each entry
            article = Article(entry.link)
            article.download()  # Download the article's content
            article.parse()     # Parse the content to extract details

            # Append the extracted data to the articles list
            articles.append({
                'title': article.title,
                'author': article.authors,
                'publish_date': article.publish_date,
                'content': article.text
            })
        except Exception as e:
            # Handle errors gracefully for problematic articles
            print(f"Error processing article: {entry.link}\n{e}")
            continue

    return articles


# RSS feed URL (BBC News in this case)
feed_url = 'http://feeds.bbci.co.uk/news/rss.xml'

# Call the function to scrape articles
articles = scrape_news_from_feed(feed_url)

# Print the extracted articles
for idx, article in enumerate(articles, 1):
    print(f"Article {idx}")
    print("Title:", article['title'])
    print("Author:", article['author'])
    print("Publish Date:", article['publish_date'])
    print("Content:", article['content'][:500], "...")  # Display the first 500 characters of the content
    print("-" * 80)
```
